```python
import matplotlib.pyplot as plt

def bresenham_line_pk(x1, y1, x2, y2):
    points = []

    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1

    dx = x2 - x1
    dy = y2 - y1

    pk = 2 * dy - dx  # Initial decision parameter

    x = x1
    y = y1

    while x <= x2:
        points.append((x, y))

        if pk < 0:

            pk = pk + 2 * dy
            x = x + 1

        else:

            pk = pk + 2 * dy - 2 * dx
            x = x + 1
            y = y + 1

    return points

# Example usage
x1_pk, y1_pk = 0, 0
x2_pk, y2_pk = 10, 6
line_points_pk = bresenham_line_pk(x1_pk, y1_pk, x2_pk, y2_pk)

# Plot the points
plt.figure(figsize=(8, 6))
plt.plot([p[0] for p in line_points_pk], [p[1] for p in line_points_pk], 'ro-')
plt.title("Bresenham's Line Algorithm (using pk)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
